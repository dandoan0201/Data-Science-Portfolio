```python
#https://www.geeksforgeeks.org/python/python-datetime-strptime-function/
import datetime
from datetime import datetime

# TOTAL_PEOPLE column??

list_time = []
list_totalpeople = []

index = 0

for item in df_test_lastentrydate_column:
    #print(len(str(item)))
    #break
    if(len(str(item)) == 16):
        #print(len(item))
        datetime_obj = datetime.strptime(item, "%m/%d/%Y %H:%M")
        time = datetime_obj.time()

        datetime_string = str(time)
        #test = time.strftime('%H:%M')
        #print(str(test))
        #print(str(time))
        #time




        #print(datetime_string)
        list_time.append(datetime_string)
        list_totalpeople.append(df_test_totalpeople_column[index])
        #break
    index = index + 1
    
list_time
```
