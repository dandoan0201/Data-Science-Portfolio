```python
# UNR_16: unemployment rate of the population aged 16 years old or older
#np.median(filtered_data['UNR_16'])

filtered_data['UNR_16']
```

```python
# UNR_16: unemployment rate of the population aged 16 years old or older
# Group by 'group' and calculate the median of 'value'
filtered_data_median_by_county = filtered_data.groupby('County_Name')['UNR_16'].median()

#filtered_data_median_by_county.values.tolist()
```

```python
#filtered_data_median_by_county.index.tolist()
```

```python
new_dataframe = {'County_Name': filtered_data_median_by_county.index.tolist(),
        'Value': filtered_data_median_by_county.values.tolist()}
new_df = pd.DataFrame(new_dataframe)

new_df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,6))
plt.bar(new_df['County_Name'], new_df['Value'])
plt.title('Unemployment Rate (Aged 16 or older)')
plt.xlabel('County Name')
plt.xticks(rotation='vertical')
plt.ylabel('Percentage (%)')
#plt.figure(figsize=(10, 8))
plt.show()
```

```python
import seaborn as sns



sns.histplot(new_df['Value'], kde=True, bins=30, color='purple')


# Labels and title
plt.xlabel('Unemployment by %')
plt.ylabel('Frequency')
plt.title('Unemployment Rate (Aged 16 or older)')
plt.grid()
plt.show()

```

```python
# ME: Median earnings expressed in 2018 dollars for the population aged 16 years old or older 
filtered_data_median_earnings_16_or_older_by_county = filtered_data.groupby('County_Name')['ME_2018'].median()
filtered_data_median_earnings_16_or_older_by_county
```

```python
new_dataframe_2 = {'County_Name': filtered_data_median_earnings_16_or_older_by_county.index.tolist(),
        'Value': filtered_data_median_earnings_16_or_older_by_county.values.tolist()}
new_df_2 = pd.DataFrame(new_dataframe_2)

new_df_2.head()
```

```python
plt.figure(figsize=(15,6))
plt.bar(new_df_2['County_Name'], new_df_2['Value'])
plt.title('Median Earnings in 2018 (Aged 16 or older)')
plt.xlabel('County Name')
plt.xticks(rotation='vertical')
plt.ylabel('Dollars ($)')
#plt.figure(figsize=(10, 8))
plt.show()
```

```python

sns.histplot(new_df_2['Value'], kde=True, bins=30, color='purple')

# Labels and title
plt.xlabel('Dollars ($)')
plt.ylabel('Frequency')
plt.title('Median Earnings in 2018 (Aged 16 or older)')
plt.grid()
plt.show()
```
